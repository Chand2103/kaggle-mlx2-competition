```python
import pandas as pd

df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")

df_train.describe()
```

```python
categorical_columns = df_train.select_dtypes(include=['object', 'category']).columns
for x in categorical_columns:
    print(x)
    print(df_train[x].values)
    print("\n")
```

```python
df_train.drop(columns=['composition_label_0', 'composition_label_1', 'publication_timestamp',
                       'creator_collective', 'composition_label_2', 'track_identifier'], inplace=True)

df_test.drop(columns=['composition_label_0', 'composition_label_1', 'publication_timestamp',
                      'creator_collective', 'composition_label_2', 'track_identifier'], inplace=True)
```

```python
df_train.describe()
```

```python
df_train.fillna(df_train.select_dtypes(include='number').mean(), inplace=True)
df_test.fillna(df_test.select_dtypes(include='number').mean(), inplace=True)
```

```python
cat_cols = df_train.select_dtypes(include=['object', 'category']).columns
df_train[cat_cols] = df_train[cat_cols].fillna('other')

df_test[cat_cols] = df_test[cat_cols].fillna('other')
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.model_selection import train_test_split


X = df_train.drop(columns=['target'])
Y = df_train['target']
X = pd.get_dummies(X, drop_first=False)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=.2)

model = XGBRegressor(objective='reg:squarederror')  # 'reg:squarederror' uses RMSE by default

# Fit the model
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:", rmse)
```

```python
df_test = pd.get_dummies(df_test, drop_first=False)

preds = model.predict(df_test)

df_sample = pd.read_csv('sample_submission.csv')

df_sample['target'] = preds

df_sample.to_csv("predictions2.csv",index=False)
```
